```python
#PIVOT allows you totransform or reshape your dataframe
# for data analysis piurpose you may want to  transform your data 
import pandas as pd
df = pd.read_csv('weather.csv')
df
```

```python
df.pivot(index='date', columns='city', values=['humidity','temperature'])
# index is what you want vertically, column is horizaontally, and values is horizonticall what you want

# IF Index contains duplicate entries, cannot reshape
```

```python
df = pd.read_csv('weather2.csv')
df
```

```python
df.pivot_table(index='city', columns='date', aggfunc='count', margins=True)
#it uses numpy functions, we can use mean(default), count, sum, mean, diff, etc
# margins gives total vertically and horizontally
```
