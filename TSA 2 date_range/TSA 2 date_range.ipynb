```python
'''
DATA RANGE:
It helps in datasets which does not have dates. We can, using this function, insert date for the dataframe.




'''
import pandas as pd
df = pd.read_csv('aapl_no_dates.csv')
df
```

```python
rng = pd.date_range(start='6/1/2017', end='6/30/2017', freq='B')
rng
```

```python
df = df.set_index(rng)
df
```

```python
%matplotlib inline
df.Close.plot()
```

```python
# to re index the dataframe we use asfreq
df.asfreq('D', method='pad')
```
