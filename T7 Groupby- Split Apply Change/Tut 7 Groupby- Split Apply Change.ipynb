```python
import pandas as pd
df = pd.read_csv('weather_by_cities.csv')
df
```

```python
g = df.groupby('city') #this will create City as key and all other information as Values for those particular key
g

# above line of code is similar to the one in  SQL
# SELECT * FROM city_data GROUP BY city 
```

```python
for city, city_df in g: # it will automatically create df for each cities which is city_df
    print(city)
    print(city_df)
    
```

```python
# to get a particular group information
g.get_group('mumbai')
```

```python
g.max() #to get max from each groups belonging to each keys which are cities
```

```python
g.mean() #to get max from each groups belonging to each keys which are cities
```

```python
# to get ALL STATISTICS at one
g.describe()
```

```python
# to plot using MATPLOTLIB
%matplotlib inline
g.plot()
```
