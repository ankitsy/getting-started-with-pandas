```python
import pandas as pd
import numpy as np
df = pd.read_csv('weather_data.csv')
df
```

```python
new_df = df.replace(-99999, np.NaN)

# if more than 1 value to be replaceds then provide list eg. below
# new_df = df.replace([-99999,-88888], np.NaN)

new_df 
```

```python
# to make replacements in COLUMNS we use dictionaries

new_df = df.replace({
    'temperature' : -99999,
    'windspeed' : -88888,
    'event' : '0'
}, np.NaN)
new_df
```

```python
#to replace multiple values like -99999 and -88888 with differnt values like NaN using mapping
new_df = df.replace({
    -99999 : 'xyz',
    -88888 : np.NaN
})
new_df
```

```python
#lets say, temperature or windspeed has values which contains words, eg. 32.0 F,  6.0 kmph
#then we use regex to remove the non-digit characters

#FOR ENTIRE DATABASE
new_df = df.replace('[A-Za-z]','',regex=True) #we inserted values we want to replace(A_Za-z) and with which we want it replaced(here we used '' which means blank)
new_df
```

```python
#FOR SPECIFIC COLOUMNS IN DATABASE: we use dictionary here
new_df = df.replace({
    'temperature': '[A_Za-z]',
    'windspeed': '[A_Za-z]'
},'',regex=True)
new_df
```

```python
#to replace LIST OF VALUES with LIST OF VALUES with mapping
new_df = df.replace(['Rain','Snow','Sunny','0'],['My','Real','Name','is'])
new_df
```
