```python
import pandas as pd
df = pd.read_csv('weather_data.csv', parse_dates=['day']) # parse_dates used because dates was a string earlier
df.set_index('day', inplace=True) 
df
```

```python
new_df = df.fillna(0) # it will fill ENTIRE DATAFRAME with 0 value
new_df
```

```python
# to change na values for specific columns we use the dictionary
new_df = df.fillna({
    'temperature' : 0,
    'windspeed' : 0,
    'event' : 'np event'
})
new_df
```

```python
# filling the na values with 0 in temperature and windspeed represents misinformation, so we use different filling methods
# we use ffill method here to fill na values with prev days value
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df = df.fillna(method="bfill", axis='columns') #bfill copying next day's value; axis column is copying values horizontally
new_df
```

```python
#limit: will copy to just next cell if limit is 1, copy two times when limit is 2 
new_df = df.fillna(method="ffill", limit=1) 
new_df
```

```python
# linear interpolate: 'linear': Ignore the index and treat the values as equally spaced..
# it is a better guess for missing values
new_df = df.interpolate() #by default method is linear here
new_df 
```

```python
new_df = df.interpolate(method='time') # Works on daily and higher resolution data to interpolate given length of interval.
new_df 
```

```python
#to DROP THE ROWS with even 1 na value
new_df= df.dropna()
new_df
```

```python
#to DROP THE ROWS with ALL na values
new_df= df.dropna(how='all')
new_df
#row for 09th january is dropped
```

```python
#to KEEP THE ROWS with atleast 1 NON na values
new_df= df.dropna(thresh=1)
new_df
```

```python
#to KEEP THE ROWS with atleast 2 NON na values

new_df= df.dropna(thresh=2)
new_df
```

```python
#insert the MISSING DATES
dt = pd.date_range('2017-01-01','2017-01-11') #created dt containing the date range
idx = pd.DatetimeIndex(dt) #created idx as DatetimeIndex containing date range
df = df.reindex(idx) #reindexed the dataframe
df
```
